```python
import seaborn as sns
import pandas as pd
import numpy.random as np
import matplotlib.pyplot as plt
```

```python
def model(inputs):
    
    #Global Paramaters

    #Amount of time to model
    time = inputs['Duration in Years']

    statics = {
        'Home Budget':inputs['Monthly Housing Budget'] * 12,
        'Home Cost': inputs["Monthly Home Cost"]*12,
        'principal_and_interest': inputs["Monthly Home Cost"] - inputs["Mortgage Extras"],
        #Total Home Value
        'original_home_value': inputs["Home Value"],
        #Initial house costs
        'house_fees': inputs["Home Purchase Fees"],
         #Mort  gage loan interest
        'interest_rate': inputs["Mortgage Interest Rate"],
        #Expected annual increase in rent
        'rent_increase': inputs["Rate of Rent Increase"],    
        #Average stock market return
        'market_return': inputs["Rate of Market Return"],
         #House Appreciation
        'home_value_increase': inputs["Rate of Home Appreciation"],
    }
    #House related parameters
    #Total Home Value
    original_home_value = inputs["Home Value"]
    #Down payment
    down_payment = inputs["Down Payment"]
    
   
    
    #Rent related parameters 
    #rent cost
    rent_cost = inputs["Rent"]   
    
    #Investment related parameters
    #Amount invested
    starting_investments = inputs["Starting Investments"]

    #Construct first line of table

    #Calculate total annual housing cost
    annual_home_cost = statics['Home Cost']

    #Calculate annual rent
    annual_rent_cost = rent_cost *12

    #Calculate starting investments
    investments = starting_investments - down_payment - statics['house_fees']

    #Calculate mortgage loan
    original_mortgage = original_home_value - down_payment
    

    years = []
    years.append({
             'home_value': original_home_value,
             'home_cost': annual_home_cost + original_home_value * .022,
             'rent_cost': annual_rent_cost,
             'mortgage': original_mortgage,
             'home_equity': down_payment,
             'investments': investments, 
             'gross': down_payment + investments,
             'Condition': inputs['Condition']
        })
    while len(years) <= time:
        years.append(one_year(years[-1], statics))
    return years
```

```python
#Run one year of the model

def one_year(previous_year, statics):


    new_home_value = previous_year['home_value'] + np.standard_gamma(statics['home_value_increase'] * 100)/80 * previous_year['home_value']
   
    #Allocate 1% of home value to repairs annually, 1.2% in expected property taxes
    new_home_cost = statics['Home Cost'] + new_home_value * .022

    new_rent_cost = previous_year['rent_cost'] + np.standard_gamma(statics['rent_increase'] * 100)/80 * previous_year['rent_cost']
    
    new_mortgage = amortization(previous_year['mortgage'], statics)
    
    new_home_equity = previous_year['home_equity'] + (previous_year['mortgage'] - new_mortgage)
   
    total_investments = previous_year['investments'] + (previous_year['investments'] * np.normal(statics['market_return'], .2, 1)) + (statics['Home Budget'] - (new_rent_cost + new_home_cost)) 
    
    new_gross = total_investments + (new_home_value - statics['original_home_value']) + new_home_equity
    
    new_year = {
             'home_value': new_home_value,
             'home_cost': new_home_cost,
             'rent_cost': new_rent_cost,
             'mortgage': new_mortgage,
             'home_equity': new_home_equity,
             'investments': total_investments, 
             'gross': new_gross[0]
        }
    return new_year
    
```

```python
def amortization(mortgage, statics):
    month = 0
    monthly_interest = statics['interest_rate']/12
    while month < 12:
        #Generate an amount paid towards interest each month
        interest_payment = mortgage * monthly_interest
        new_equity = statics['principal_and_interest'] - interest_payment
        mortgage = mortgage - new_equity
        month += 1
        
    return mortgage
```

```python
#Controller
def controller(inputs):

    repetitions= inputs['Replicates']
    repetition = []
    while repetitions  > 0:
        repetition.append(model(inputs))
        repetitions -= 1
    return repetition
```

```python
def line_plot(reps):
    sns.set_theme()
    plt.figure(figsize=(16, 6))
    gross = pd.DataFrame(reps).applymap(lambda x: x['gross']).T
    gross.index.names = ['Year']
    gross_melty = pd.melt(gross, ignore_index=False)
    sns.lineplot(data=gross, legend=False, alpha = 0.4)
    sns.lineplot(data=gross_melty, x='Year', y = 'value', ci='sd')
    plt.savefig(reps[0][0]['Condition'])
    
```

```python
def multi_line(reps):
    gross = pd.DataFrame(reps).applymap(lambda x: x['gross']).T
    sns.lineplot(data=gross, legend=False, alpha = 0.4).get_figure()
```

```python
#Pass violin_compare a list of lists, where the key is a label and the content is an output from running controller
def violin_compare(listOfLists):
    toPandas = {
        'Condition': [],
        'Gross': [],
    }
    
    for element in listOfLists:
        element[1] =  [repetition[-1]["gross"] for repetition in element[1]].copy()
        toPandas['Gross'] += element[1]
        toPandas['Condition'] += [element[0]]*len(element[1])
        print(f'{element[0]} Average: {sum(element[1])/len(element[1])}')
    df = pd.DataFrame(toPandas)
    #return df
    sns.set(rc = {'figure.figsize':(15,8)})
    sns.violinplot(x='Condition', y='Gross', data=df)
    
    
def batch_violin(listOfLists):
    run_list = []
    for element in listOfLists:
        run_list.append([element[0] ,controller({param_names[i]: element[i] for i in range(len(element))})])
    violin_compare(run_list)
```
